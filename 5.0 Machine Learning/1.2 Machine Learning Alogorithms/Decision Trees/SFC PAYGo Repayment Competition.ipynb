```python
merged = pd.merge(metadata, train, on='ID')
merged2 = pd.merge(metadata, test, on='ID')
merged.head(2)
```

```python
for i in range(21):
    merged.PaymentsHistory = merged.PaymentsHistory.str[1:-2]
    merged['PaymentsHistory{}'.format(i)] = merged.PaymentsHistory.str.split(',', expand=True)[i]
    merged2.PaymentsHistory = merged2.PaymentsHistory.str[1:-2]
    merged2['PaymentsHistory{}'.format(i)] = merged2.PaymentsHistory.str.split(',', expand=True)[i]
```

```python
merged.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
gender_encoded = le.fit_transform(merged['MainApplicantGender'])
region = le.fit_transform(merged['Region'])
gender_encoded2 = le.fit_transform(merged2['MainApplicantGender'])
region2 = le.fit_transform(merged2['Region'])

merged['MainApplicantGender'] = gender_encoded.astype('float64')
merged['Region'] = region.astype('float64')
merged2['MainApplicantGender'] = gender_encoded2.astype('float64')
merged2['Region'] = region2.astype('float64')
```

```python
training = merged[['AccessoryRate', 'RatePerUnit', 'DaysOnDeposit', 'Term', 'TotalContractValue', 'MainApplicantGender', 
                   'Region', 'PaymentsHistory0', 'PaymentsHistory1', 'PaymentsHistory2', 'PaymentsHistory3', 'PaymentsHistory4',
                   'm1', 'm2', 'm3', 'm4', 'm5', 'm6']]
#                  'PaymentsHistory5', 'PaymentsHistory6', 'PaymentsHistory7', 'PaymentsHistory8','PaymentsHistory9', 
#                  'PaymentsHistory10']]

testing =  merged2[['AccessoryRate', 'RatePerUnit', 'DaysOnDeposit', 'Term', 'TotalContractValue', 'MainApplicantGender', 
                    'Region', 'PaymentsHistory0', 'PaymentsHistory1', 'PaymentsHistory2', 'PaymentsHistory3', 
                    'PaymentsHistory4']]
```

```python
training.info()
```

```python
testing.info()
```

```python
training = training.astype('float64')
training
testing = testing.astype('float64')
testing
```

```python
training.fillna(0, inplace=True)
```

```python
testing.fillna(0, inplace=True)
```

```python
X = training.drop(['m1', 'm2', 'm3', 'm4', 'm5', 'm6'], 1)
y1 = training.m1
y2 = training.m2
y3 = training.m3
y4 = training.m4
y5 = training.m5
y6 = training.m6

testing = testing
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y1, test_size=0.2, random_state=20)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y2, test_size=0.2, random_state=20)
X_train3, X_test3, y_train3, y_test3 = train_test_split(X, y3, test_size=0.2, random_state=20)
X_train4, X_test4, y_train4, y_test4 = train_test_split(X, y4, test_size=0.2, random_state=20)
X_train5, X_test5, y_train5, y_test5 = train_test_split(X, y5, test_size=0.2, random_state=20)
X_train6, X_test6, y_train6, y_test6 = train_test_split(X, y6, test_size=0.2, random_state=20)
```

```python
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
from catboost import CatBoostRegressor
```

```python
model1 = CatBoostRegressor(iterations=350, learning_rate=0.01, depth=8, l2_leaf_reg=0.6, subsample=1.0, max_depth=None)
model2 = CatBoostRegressor(iterations=350, learning_rate=0.01, depth=8, l2_leaf_reg=0.6, subsample=1.0, max_depth=None)
model3 = CatBoostRegressor(iterations=350, learning_rate=0.01, depth=8, l2_leaf_reg=0.6, subsample=1.0, max_depth=None)
model4 = CatBoostRegressor(iterations=350, learning_rate=0.01, depth=8, l2_leaf_reg=0.6, subsample=1.0, max_depth=None)
model5 = CatBoostRegressor(iterations=350, learning_rate=0.01, depth=8, l2_leaf_reg=0.6, subsample=1.0, max_depth=None)
model6 = CatBoostRegressor(iterations=350, learning_rate=0.01, depth=8, l2_leaf_reg=0.6, subsample=1.0, max_depth=None)

model1a = GradientBoostingRegressor(learning_rate=0.1, n_estimators=200, subsample=1.0, criterion='friedman_mse', max_depth=3)
model2a = GradientBoostingRegressor(learning_rate=0.1, n_estimators=200, subsample=1.0, criterion='friedman_mse', max_depth=3)
model3a = GradientBoostingRegressor(learning_rate=0.1, n_estimators=200, subsample=1.0, criterion='friedman_mse', max_depth=3)
model4a = GradientBoostingRegressor(learning_rate=0.1, n_estimators=200, subsample=1.0, criterion='friedman_mse', max_depth=3)
model5a = GradientBoostingRegressor(learning_rate=0.1, n_estimators=200, subsample=1.0, criterion='friedman_mse', max_depth=3)
model6a = GradientBoostingRegressor(learning_rate=0.1, n_estimators=200, subsample=1.0, criterion='friedman_mse', max_depth=3)
```

```python
model1.fit(X_train1, y_train1)
model2.fit(X_train2, y_train2)
model3.fit(X_train3, y_train3)
model4.fit(X_train4, y_train4)
model5.fit(X_train5, y_train5)
model6.fit(X_train6, y_train6)

model1a.fit(X_train1, y_train1)
model2a.fit(X_train2, y_train2)
model3a.fit(X_train3, y_train3)
model4a.fit(X_train4, y_train4)
model5a.fit(X_train5, y_train5)
model6a.fit(X_train6, y_train6)
```

```python
preds1 = model1.predict(X_test1)
preds2 = model2.predict(X_test2)
preds3 = model3.predict(X_test3)
preds4 = model4.predict(X_test4)
preds5 = model5.predict(X_test5)
preds6 = model6.predict(X_test6)

preds1a = model1a.predict(X_test1)
preds2a = model2a.predict(X_test2)
preds3a = model3a.predict(X_test3)
preds4a = model4a.predict(X_test4)
preds5a = model5a.predict(X_test5)
preds6a = model6a.predict(X_test6)
```
